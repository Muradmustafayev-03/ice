```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('ICE.csv', sep='  ', names=['Year',
                                             'January',
                                             'February',
                                             'March',
                                             'April',
                                             'May',
                                             'June',
                                             'July',
                                             'August',
                                             'September',
                                             'October',
                                             'November',
                                             'December'])
df
```

```python
x_axis = []
for row in df.values:
    for month in range(12):
        x_axis.append(row[0] + (month + 0.5) / 12)
```

```python
y_axis = df.drop(columns='Year').values.flatten()
```

```python
x_axis_smooth = [np.mean(array) for array in np.split(np.array(x_axis), len(x_axis) / 12)]
print(x_axis_smooth)
```

```python
y_axis_smooth = [np.mean(array) for array in np.split(np.array(y_axis), len(y_axis) / 12)]
print(y_axis_smooth)
```

```python
fig, ax0 = plt.subplots()
ax0.plot(x_axis, y_axis, label='monthly')
ax0.plot(x_axis_smooth, y_axis_smooth, label='smoothed')
ax0.legend(loc='upper right')
ax0.set_ylabel('Ice covered area [mln sq km]')
ax0.set_xlabel('Year')
ax0.set_title('Observations, North Ocean')
```

```python
means = np.array([np.ones(12) * mean for mean in y_axis_smooth])
seasonal_osc = df.drop(columns='Year').values - means
monthly_changes_avg = [np.mean(months) for months in seasonal_osc.transpose()]
monthly_changes_avg
```

```python
trend = LinearRegression()
trend.fit(np.array(x_axis_smooth).reshape(-1, 1), np.array(y_axis_smooth).reshape(-1, 1))
trend.score(np.array(x_axis_smooth).reshape(-1, 1), np.array(y_axis_smooth).reshape(-1, 1))
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(np.array(x_axis).reshape(-1, 1), trend.predict(np.array(x_axis).reshape(-1, 1)))
ax1.plot(x_axis_smooth, y_axis_smooth)
```

```python
def which_month(year):
    return int(np.round((year - int(year)) * 12 - 0.5))
```

```python
def predict(x):
    return [sum(x) for x in zip(trend.predict(np.array(x).reshape(-1, 1)), np.array([monthly_changes_avg[which_month(year)] for year in x]))]
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(x_axis, y_axis, label='real')
ax2.plot(x_axis, predict(x_axis), label='predicted')
```
